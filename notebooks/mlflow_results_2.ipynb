#### Sharp problem
* What: In `base` run, CNN trained on SHARP withseed 3 has particularly bad performance.
* Why: AUC is a bad metric. The step 494 (end of epoch 8) has higher validation AUC than its neighboring epochs (not very distinguishable, 0.994 vs 0.986 and 0.989), but other metrics are lower (e.g., 0.702 vs 0.840 and 0.883). Looking at SSP, the curve significantly tilted leftwards, indicating a significant amount of underforecasting.
* This also suggests using an optimized TSS is a dangerous idea because it can select a under- or over-forecasting model that is not robust to the threshold.
    * It is very much possible that the test SSP is a bit different from the validation SSP, and using the threshold fitted on the validation set may lead to a performance much lower than expected. This is due to the sensitivity of SSP to the threshold.
    * Thresholding on the test set may not be very useful in practice, as we are not allowed to give a prediction of a test instance unless we know the entire test set.
* Why: AUC is not consistent with others is because it is the area under the uniform measure of false alarm rate, which is rarely the case in practice where false alarm rate is selected to be close to zero. AUC gives too much weight on the part where false alarm is high.
    * Area Under SSP may be a good idea. It is the area using the uniform measure of threshold.

* Conclusion 1 & 2 are in conflict.
    * The difference may not be due to how well data itself is related to flares. It could also be the way we fuse the data. But if there is something wrong, for parameters. E.g., it could be that when adding SHARP parameters to the training set, we do not normalize correctly, but when the other way around, we normalized SMARP correctly. We did normalizaiton separately for SHARP and SMARP. 
* Inspection: Check the distributions *after* transformation.
    * But both SMARP and SHARP should have mean 0 and std 1 after standardization.
    * Does it make sense to normalization the two datasets separately?
        * One cycle could have fewer flares, and parameters are smaller. Normalizing separately artificially increase the activity of this cycle.
        * Normalizing separately assumes that the data for different solar cycles have the same mean and same std for the "true value" of every keyword.
            * The keyword parameters are measurements of the "true value". It may also depends on the unit.
            * If the first two moments are different, then separated normalization will create a bias to one of the dataset, making its values larger or more variable.
        * RUS doesn't make a difference. We are not computing mean and std with RUS-ed data.
        * The problem is, we are not able to recover the "true value".
            * We can derive the transformation using the co-temporal measurements. This should generalize well.
            * We can convert MEANGBL to the same physical unit. But this should be accounted for in the linear transformation derived above.
            * This seems a valid approach. Why we decided not to use it?
                * Because we decided to normalize separately. In that case, performing a transformation to one of them will have no effect after normalization of that dataset.
                * However, the transformation certainly makes a difference if we perform a global normalization. Indeed, the normalization should be performed after correction/calibration/fusion. You have messed up the order.

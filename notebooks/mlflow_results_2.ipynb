#### Sharp problem
* What: In `base` run, CNN trained on SHARP withseed 3 has particularly bad performance.
* Why: AUC is a bad metric. The step 494 (end of epoch 8) has higher validation AUC than its neighboring epochs (not very distinguishable, 0.994 vs 0.986 and 0.989), but other metrics are lower (e.g., 0.702 vs 0.840 and 0.883). Looking at SSP, the curve significantly tilted leftwards, indicating a significant amount of underforecasting.
* This also suggests using an optimized TSS is a dangerous idea because it can select a under- or over-forecasting model that is not robust to the threshold.
    * It is very much possible that the test SSP is a bit different from the validation SSP, and using the threshold fitted on the validation set may lead to a performance much lower than expected. This is due to the sensitivity of SSP to the threshold.
    * Thresholding on the test set may not be very useful in practice, as we are not allowed to give a prediction of a test instance unless we know the entire test set.
* Why: AUC is not consistent with others is because it is the area under the uniform measure of false alarm rate, which is rarely the case in practice where false alarm rate is selected to be close to zero. AUC gives too much weight on the part where false alarm is high.
    * Area Under SSP may be a good idea. It is the area using the uniform measure of threshold.

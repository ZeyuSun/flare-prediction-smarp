# Conv filters

```python
import matplotlib.pyplot as plt
from dashboard_helper import get_learner
```

```python
learner = get_learner('sharp', 0, 'CNN', experiment_name='CNN', run_name='More_epochs')
```

```python
%matplotlib inline
from operator import attrgetter
filters = attrgetter('convs.conv1')(learner.model).weight.detach().cpu().numpy()
plt.figure(figsize=(10,10))
for i in range(8):
    for j in range(8):
        plt.subplot(8, 8, i*8+j+1)
        plt.imshow(filters[i*8+j, 0, 0], cmap='gray')
        plt.axis('off')
plt.show()
```
